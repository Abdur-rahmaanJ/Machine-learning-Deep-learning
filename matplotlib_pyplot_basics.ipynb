# matplotlib

import pyplot

```python
import matplotlib.pyplot as plt
```

simple plot

```python
list = [1, 2, 3, 4]
plt.plot(list)
plt.show()
```

plotting by two cordinates x, y

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 10, 12, 22]
plt.plot(x, y)
plt.show()
```

with green circles

```python
# go
x = [1, 2, 3, 4, 5]
y = [2, 4, 10, 12, 22]
plt.plot(x, y, 'go')
plt.show()
```

green squares

```python
# gs
x = [1, 2, 3, 4, 5]
y = [2, 4, 10, 12, 22]
plt.plot(x, y, 'gs')
plt.show()
```

green triangles

```python
# g^
x = [1, 2, 3, 4, 5]
y = [2, 4, 10, 12, 22]
plt.plot(x, y, 'g^')
plt.show()
```

dashed

```python
# g--
x = [1, 2, 3, 4, 5]
y = [2, 4, 10, 12, 22]
plt.plot(x, y, 'g--')
plt.show()
```

line width

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 10, 12, 22]
plt.plot(x, y, linewidth=6.0)
plt.show()
```

multiple plots

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 10, 12, 22]
x2 = [1, 3, 5, 7, 9]
y2 = [2, 4, 6, 10, 18]
plt.plot(x, y, 'ro', x2, y2, 'bo')
plt.show()
```

### adding labels

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 10, 12, 22]
plt.plot(x, y, linewidth=6.0)
plt.xlabel('time', fontsize=18)
plt.ylabel('height', fontsize=18)
plt.show()
```

legend

```python
import matplotlib.patches as mpatches

red_patch = mpatches.Patch(color='red', label='The red data')
plt.legend(handles=[red_patch])

plt.show()
```

## title

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 10, 12, 22]
plt.plot(x, y, linewidth=6.0)
plt.xlabel('time', fontsize=18)
plt.ylabel('height', fontsize=18)
plt.title('demo data')
plt.show()
```

## grid

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 10, 12, 22]
plt.plot(x, y, linewidth=6.0)
plt.xlabel('time', fontsize=18)
plt.ylabel('height', fontsize=18)
plt.grid(True)
plt.show()
```
